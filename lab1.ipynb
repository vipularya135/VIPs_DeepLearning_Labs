sigmoid

```python
import numpy as np

def basic_sigmoid(x):

    s = 1 / (1 + np.exp(-x))

    return s
```

```python
basic_sigmoid(1)
```

```python
import math

def math_sigmoid(x):

    s = 1 / (1 + math.exp(-x))

    return s
```

```python
math_sigmoid(1)
```

sigmoid derivative

```python
#sigmoid derivative

def sigmoid_derivative(x):

    s = math_sigmoid(x)
    ds = s * (1 - s)

    return ds
```

```python
sigmoid_derivative(1)
```

```python
sigmoid_derivative(2)
```

```python
def imagetovector(image):

    v = image.reshape(image.shape[0] * image.shape[1] * image.shape[2], 1)

    return v
```

normalising data/ rows

```python
def normalize(x):

    x_norm = np.linalg.norm(x, axis = 1, keepdims = True)
    x = x / x_norm

    return x
```

```python
x1 = np.array([[0,1000,4], [2,6,4]])
```

```python
normalize(x1)
```

softmax

```python
def softmax(x):

    x_exp = np.exp(x)
    x_sum = np.sum(x_exp, axis = 1, keepdims = True)
    s = x_exp / x_sum

    return s
```

```python
softmax(x1)
```

vectorization

```python
import time
#dot product
x1 = [9, 2, 5, 0, 0, 7, 5, 0, 0, 0, 0, 9, 2, 5, 0, 0]
x2 = [9, 2, 2, 9, 0, 9, 2, 5, 0, 0, 9, 2, 5, 0, 0, 9]
tic = time.process_time()
dot = 0
for i in range(len(x1)):
    dot+= x1[i]*x2[i]
toc = time.process_time()
print ("dot = " + str(dot) + "\n ----- Computation time = " + str(1000*(toc - tic)) + "ms")
```

```python
# outer product
ticc = time.process_time() # Call the function and assign the result to ticc
outer = np.zeros((len(x1),len(x2)))
for i in range(len(x1)):
    for j in range(len(x2)):
        outer[i,j] = x1[i]*x2[j]
tocc = time.process_time()
print ("dot = " + str(dot) + "\n ----- Computation time = " + str(1000*(tocc - ticc)) + "ms")
```
